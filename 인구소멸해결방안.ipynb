##2017 ~ 2021년 소멸위험지수 계산

20-39세 여성인구 파일 읽어들임

```python
import pandas as pd

women = pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/RawData/women17_21.xlsx')
women
```

여성인구 합계

```python
women['sum'] = women[20] + women[25] + women[30] +women[35]
women = women.drop([20,25,30,35, '성별'], axis=1)
women
```

65세 고령인구 

```python
old = pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/RawData/old17-21.xlsx')
old
```

65세 이상 고련인구 합계

```python
old['sum'] = old[65] + old[70] + old[75] +old[80]
old = old.drop([65,70,75,80], axis=1)
old
```

소멸지수 구한 데이터 프레임

```python
score = old
score['소멸지수'] = women['sum'] / old['sum']
score = score.drop('sum', axis=1)
score.columns = ['시도', '시군구', '연도', '소멸지수']
score.round(3)
```

시군구앞 공백 제거

```python
score['시군구'] = score['시군구'].str.lstrip()
```

```python
score.to_excel('score.xlsx')
```

지역의 개수를 이동자수 기준으로 맞추기 위해서 merge진행

```python
move = pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/data/region%20runoff%20rate.xlsx', index_col=0)
move.columns = ['시도', '시군구', '연도', '유출률']
move
```

```python
lastscore = pd.merge(move, score, on=['시도', '시군구', '연도'], how='left')
lastscore = lastscore.drop('유출률', axis=1)
lastscore
```

최종 소멸지수 파일 다운로드

```python
lastscore.to_excel('소멸지수ㄹㅇ이게마지막진짜로최종에끝자락더이상안할꺼임수정안해버전.xlsx')
```

##합계출산율

```python
import pandas as pd

df = pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/RawData/TFR_RawData.xlsx')
```

```python
df17 = df.drop(['2018','2019','2020','2021'],axis=1)
df17['연도'] = 2017
df17 = df17.rename(columns={'2017':'합계출산율'})

df18 = df.drop(['2017','2019','2020','2021'],axis=1)
df18['연도'] = 2018
df18 = df18.rename(columns={'2018':'합계출산율'})

df19 = df.drop(['2017','2018','2020','2021'],axis=1)
df19['연도'] = 2019
df19 = df19.rename(columns={'2019':'합계출산율'})

df20 = df.drop(['2017','2018','2019','2021'],axis=1)
df20['연도'] = 2020
df20 = df20.rename(columns={'2020':'합계출산율'})

df21 = df.drop(['2017','2018','2019','2020'],axis=1)
df21['연도'] = 2021
df21 = df21.rename(columns={'2021':'합계출산율'})
```

```python
fertility_df = pd.concat([df17, df18, df19, df20, df21])
fertility_df = fertility_df.reindex(columns=['시도', '시군구', '연도', '합계출산율'])

#fertility_df.dtypes # 타입확인
#fertility_df = fertility_df.astype({'연도':'int32'}) #연도 타입 float64-> int 변경

fertility_df = fertility_df.sort_values(by=['시도', '시군구', '연도'], axis=0)
fertility_df = fertility_df.reset_index(drop=True)
fertility_df

#256개의 행정구역 * 5(17년~21년) = 1280개
```

```python
#지역별 유출율
ror_df=pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/data/region%20runoff%20rate.xlsx',index_col=0)
#ror_df = ror_df.round(3)
ror_df
```

```python
fertility_df = pd.merge(ror_df,fertility_df, on=['시도','시군구','연도'],how='left')

fertility_df = fertility_df.drop(['유출률'],axis=1)

fertility_df
```

```python
fertility_df.to_excel('시군구_합계출산율2017~2021(real최종).xlsx')
```
