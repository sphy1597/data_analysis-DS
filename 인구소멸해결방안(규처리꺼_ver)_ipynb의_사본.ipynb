합계출산율과 지방소멸의 연관성

2021년 시군구별 소멸위험지수 계산

```python
import pandas as pd

women = pd.read_excel('/content/여성17_21.xlsx')
women
```

```python
women['sum'] = women[20] + women[25] + women[30] +women[35]
women = women.drop([20,25,30,35, '성별'], axis=1)
women
```

```python
old = pd.read_excel('/content/고령17_21.xlsx')
old
```

```python
old['sum'] = old[65] + old[70] + old[75] +old[80]
old = old.drop([65,70,75,80], axis=1)
old
```

```python
score = old
score['소멸지수'] = women['sum'] / old['sum']
score = score.drop('sum', axis=1)
score.columns = ['시도', '시군구', '연도', '소멸지수']
score
```

```python
score.to_excel('소멸지수ver1.xlsx')
```

```python
score = pd.read_excel('/content/소멸지수ver2.xlsx', index_col=0)
```

```python
move = pd.read_excel('/content/지역별 유출률 2017_2021 최종본.xlsx', index_col=0)
move.columns = ['시도', '시군구', '연도', '유출률']
move
```

```python
lastscore = pd.merge(move, score, on=['시도', '시군구', '연도'], how='left')
lastscore = lastscore.drop('유출률', axis=1)
lastscore
```

```python
lastscore.to_excel('소멸지수ㄹㅇ이게마지막진짜로최종에끝자락더이상안할꺼임수정안해버전.xlsx')
```
