# 합계 유출률 최종 !!!!

```python
import pandas as pd
df1 = pd.read_excel('https://github.com/sphy1597/data_analysis-DS/raw/main/RawData/city%20number%20of%20movers.xlsx')
```

```python
df1
```

```python
df=df1.replace(0, pd.NA)
```

```python
df=df.dropna(axis=0)
df
```

```python
r=pd.DataFrame()
```

```python
s = pd.DataFrame()
s['전입률'] = df['총전입 (명)'].apply(lambda x: float(x))
s['전출률'] = df['총전출 (명)'].apply(lambda x: float(x))
s['전체 값'] = s['전입률'] - s['전출률']
s['유출률'] = s['전체 값']/s['전출률'] 
r['2021']= s['유출률']
```

```python
r
```

```python
df['유출률'] = r['2021']
df = df.drop(['총전입 (명)','총전출 (명)'], axis = 1)
```

```python
df = df.round(3)
df = df.reset_index(drop=True)
df.columns = ['시도','시군구','연도','유출률']
df['시군구']= df['시군구'].str.strip()
for i in range(1145):
  df['시군구'].iat[i] = df['시군구'].iat[i].replace('  ', '')
```

```python
df
```

```python
df.to_excel("지역별 유출률 2017~2021 최종본.xlsx")
```
