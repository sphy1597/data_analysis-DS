합계출산율과 지방소멸의 연관성

2021년 시군구별 소멸위험지수 계산

```python
import pandas as pd

data = pd.read_csv('/content/시군구별 인구수.csv', encoding='cp949')
data = data.drop(['계', '80 - 84세', '85세 이상', '85 - 89세', '90 - 94세', '95 - 99세', '100세 이상'], axis = 1)
```

```python
data
```

```python
score = pd.DataFrame()
score['신군구'] = data['행정구역(시군구)별']
score['성별'] = data['성별']
score['시점'] = data['시점']

score['20 - 39세'] = data['20 - 24세'] + data['25 - 29세'] + data['30 - 34세'] + data['35 - 39세']
score['65세 이상'] = data['65 - 69세'] + data['70 - 74세'] + data['75 - 79세'] + data['80세 이상']
score['소멸지수'] = score['20 - 39세'] / score['65세 이상']

score
```

```python
score.to_csv('소멸지수.csv')
```
