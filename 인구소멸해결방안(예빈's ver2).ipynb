합계출산율과 지방소멸의 연관성

2021년 시군구별 소멸위험지수 계산

```python
import pandas as pd

data = pd.read_csv('/content/시군구별 인구수.csv', encoding='cp949')
```

```python
data
```

```python
# 전국 데이터는 삭제
df_sample = data.drop(index=[0, 1], axis=0)
df_sample
```

```python
# 합계 인구수 데이터 출력
df_sample_1 = data[data['성별'].isin(["계"])]
df_sample_1
```

```python
df_sample_1['65세이상'] = df_sample_1['65 - 69세']+df_sample_1['70 - 74세']+df_sample_1['75 - 79세']+df_sample_1['80 - 84세']+df_sample_1['85 - 89세']+df_sample_1['90 - 94세']+df_sample_1['95 - 99세']+df_sample_1['100세 이상']

df_sample_1
```

```python
# 여성 인구수 데이터만 추출
df_sample_w = data[data['성별'].isin(["여자"])]
df_sample_w
```

```python
df_sample_w['20 - 39세'] = df_sample_w['20 - 24세']+df_sample_w['25 - 29세']+df_sample_w['30 - 34세']+df_sample_w['35 - 39세']
df_sample_w
```

```python
df_sample_w['소멸지수'] = df_sample_w['20 - 39세']/df_sample_1['65세이상']

df_sample_w
```

## 행추가 - 소멸지수

```python
import pandas as pd
```

```python
df1 = pd.read_excel('/content/소멸지수.xlsx', index_col=0) 
df2 = pd.read_excel('/content/지방소멸위험지수 2017_2020.xlsx', index_col=0)
```

```python
df = df1[df1["성별"]=="계"]
df 
```

```python
# df = pd.DataFrame(df2)

new_df = pd.concat([df, df2])
new_df 
```

```python
df_sample = new_df.drop(columns=['성별', '시도'], axis=1)
# df_sample
```

```python
df_sample = df_sample.sort_values(by=['시군구', '연도'], axis=0)
```

```python
df_sample = df_sample.reset_index(drop=True)
df_sample
```

```python
df_sample.to_csv('지방소멸위험지수2017-2021.csv')
```
