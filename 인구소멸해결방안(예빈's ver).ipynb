합계출산율과 지방소멸의 연관성

2021년 시군구별 소멸위험지수 계산

```python
import pandas as pd

data = pd.read_csv('https://github.com/sphy1597/data_analysis-DS/raw/main/source/population.csv')
```

```python
data
```

```python
# 전국 데이터는 삭제
df_sample = data.drop(index=[0, 1], axis=0)
# 65세 이상 column 추가
df_sample['65세이상'] = df_sample['65 - 69세']+df_sample['70 - 74세']+df_sample['75 - 79세']+df_sample['80 - 84세']+df_sample['85 - 89세']+df_sample['90 - 94세']+df_sample['95 - 99세']+df_sample['100세 이상']
# 20-39 세 여성
df_sample['20 - 39세'] = df_sample['20 - 24세']+df_sample['25 - 29세']+df_sample['30 - 34세']+df_sample['35 - 39세']

df_sample
```

```python
# 합계 인구수 데이터 출력
df_sample_1 = data[data['성별'].isin(["계"])]
df_sample_1
```

```python
df_sample_1['65세이상'] = df_sample_1['65 - 69세']+df_sample_1['70 - 74세']+df_sample_1['75 - 79세']+df_sample_1['80 - 84세']+df_sample_1['85 - 89세']+df_sample_1['90 - 94세']+df_sample_1['95 - 99세']+df_sample_1['100세 이상']

df_sample_1
```

```python
# 여성 인구수 데이터만 추출
df_sample_w = data[data['성별'].isin(["여자"])]
df_sample_w
```

```python
df_sample_w['20 - 39세'] = df_sample_w['20 - 24세']+df_sample_w['25 - 29세']+df_sample_w['30 - 34세']+df_sample_w['35 - 39세']
df_sample_w
```

```python
df_sample_w['소멸지수'] = df_sample_w['20 - 39세']/df_sample_1['65세이상']

df_sample_w
```
